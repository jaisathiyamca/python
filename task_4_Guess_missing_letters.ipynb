```python
import random

def findMissingLetter():
    randomChoice=random.choice(['dad','mom','friend','baby','uncle'])
    indicesToRemove=random.randrange(0,len(randomChoice))
    
    convertWordintoList=list(randomChoice)
    letterToFind=convertWordintoList[indicesToRemove]
    convertWordintoList[indicesToRemove]='_'

    print(f"Find missing letter in the word: {''.join(convertWordintoList)}")
    for i in range(5):
        userInput=input('Enter Missing Letter:')
        if(len(userInput)>1):
            print('Invalid input. kindly enter only one character')
        elif letterToFind==userInput:
            break;
        else:
            print('Invalid character')
    if(letterToFind==userInput):
        print('congrats!. You have found missing character')
    else:
        print(f'You have reached maximum limit. Correct word is {randomChoice}')
            
findMissingLetter()
```
