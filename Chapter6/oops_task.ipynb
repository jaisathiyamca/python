## To excercise OOPS concept Abstraction,polymorphism,inheritance

## 1. Abstraction

```python
from abc import ABC,abstractmethod
```

```python
class Car(ABC):
    def __init__(self,brand):
        self.brand=brand
    
    @abstractmethod
    def startEngine(self):
        pass
    
    @abstractmethod
    def stopEngine(self):
        pass
```

## 2. Inheritance

```python
class PetrolCar(Car):
    def __self__(self,brand):
        super().__init__(brand)
        self.brand=brand
    def startEngine(self):
        print(f'Petrol car {self.brand} engine started')
    def stopEngine(self):
        print(f'Petrol car {self.brand} engine stopped')
    
    
        
```

```python
maruti_ciaz=PetrolCar('Maruti')
maruti_ciaz.startEngine()
```

```python
class EvCar(Car):
    def __self__(self,brand):
        super().__init__(brand)
        self.brand=brand
        
    def startEngine(self):
        print(f'Motor {self.brand} started ')
        
    def stopEngine(self):
        print(f'Motor {self.brand} stoped')
```

```python
cars =[PetrolCar('Maruti Riz'),EvCar('MG comet'),]
for c in cars:
    c.startEngine()
```

###  3. Polymorphism

```python
class Car:
    def __init__(self,brand ,model):
        self.brand=brand
        self.model=model
        
    def startEngine(self):
        print(f'Generic Car {self.brand} {self.model} Engine started ')
```

```python
class PetrolCar(Car):
    def __init__(self,brand,model):
        self.brand=brand
        self.model=model
        
    def startEngine(self):
        print(f'Petrol car {self.brand} {self.model} engine started')
```

```python
genCar=Car('Maruti','zen')
tiago=PetrolCar('TATA','Tiago')
genCar.startEngine()
tiago.startEngine()
```
